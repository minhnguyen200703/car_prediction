## 1.2 Handle Missing Values

```python
# Check for missing values
print("Missing values per column:")
print(df_car.isnull().sum())

# Drop rows with missing essential values (e.g., price or mileage_v2)
df_car = df_car.dropna(subset=['price','model'])
```

```python
# Step 1: Calculate Mode Values for Each Model
mode_origin = df_car.groupby('model')['origin'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
mode_type = df_car.groupby('model')['type'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)
mode_seats = df_car.groupby('model')['seats'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else np.nan)

# Step 2: Fill Missing or Invalid Values in Bulk
# Map the mode values for each model
df_car.loc[:, 'origin'] = df_car['origin'].fillna(df_car['model'].map(mode_origin))
df_car.loc[:, 'type'] = df_car['type'].fillna(df_car['model'].map(mode_type))

# Handle seats with both missing and invalid (-1) values
df_car.loc[:, 'seats'] = df_car.apply(
    lambda row: mode_seats[row['model']] if (pd.isnull(row['seats']) or row['seats'] == -1) else row['seats'], axis=1
)

# Step 3: Fallback to Overall Mode if Model Mapping Fails
# Fallback for origin
df_car.loc[:, 'origin'].fillna(df_car['origin'].mode()[0], inplace=True)

# Fallback for type
df_car.loc[:, 'type'].fillna(df_car['type'].mode()[0], inplace=True)

# Fallback for seats
df_car.loc[:, 'seats'].replace(-1, np.nan, inplace=True)  # Treat -1 as missing
df_car.loc[:, 'seats'].fillna(df_car['seats'].mode()[0], inplace=True)

# Fill missing gearbox and fuel with mode
df_car.loc[:, 'gearbox'] = df_car['gearbox'].fillna(df_car['gearbox'].mode()[0])
df_car.loc[:, 'fuel'] = df_car['fuel'].fillna(df_car['fuel'].mode()[0])
```

### 1.3 Observation & Domain Knowledge<a id="observation-and-domain-knowledge"></a>


### Observation & Domain Knowledge

The objective for this model is to predict the **price of used cars** based on the various features of the vehicles themselves. This can help car sellers and buyers understand the factors that influence a car's market value and make data-driven decisions.

The column we are predicting is `price` (numerical), which indicates the market price of the car in VND (Vietnamese Dong).

---

### **Features**
Based on the provided dataset, the following features will be used to predict the target:

| Feature              | Data Type | Description                                                                       | Acceptable Range                                                   |
|-----------------------|-----------|-----------------------------------------------------------------------------------|--------------------------------------------------------------------|
| `id`                 | int       | Unique identifier for the car listing                                             | Unique integers                                                   |
| `list_id`            | int       | Unique identifier for the listing                                                 | Unique integers                                                   |
| `list_time`          | datetime  | Timestamp of the listing                                                          | Unix timestamp format                                              |
| `manufacture_date`   | int       | Year of manufacture for the car                                                   | Integer (e.g., 1980–2023)                                         |
| `brand`              | string    | Brand of the car                                                                  | Examples: Toyota, Honda, Kia, Hyundai                             |
| `model`              | string    | Model of the car                                                                  | Examples: Vios, Civic, Accent                                      |
| `origin`             | string    | Country of origin                                                                | Examples: Japan, USA, Vietnam, Korea                               |
| `type`               | string    | Type of car                                                                       | Examples: Sedan, SUV, Minivan, Hatchback                          |
| `seats`              | int       | Number of seats in the car                                                        | Typical values: 4, 5, 7, 8                                        |
| `gearbox`            | string    | Gearbox type                                                                      | Examples: MT (Manual), AT (Automatic)                             |

| `color`              | string    | Exterior color of the car                                                         | Examples: White, Black, Grey, Red                                 |
| `mileage_v2`         | float     | The distance the car has been driven                                              | Kilometer values, typically greater than 0                        |
| `condition`          | string    | The condition of the car                                                          | Examples: New, Used                                               |

---

### **Additional Notes on Key Features**
1. **`mileage_v2` (Mileage)**: 
   - Mileage is expected to have a negative correlation with the car price. Higher mileage usually reduces the price of a car as it indicates more wear and tear.

2. **`manufacture_date` (Year of Manufacture)**:
   - Newer cars are expected to have higher prices compared to older models. The difference in price can vary significantly based on the brand and model's reputation.

3. **`brand` and `model`**:
   - These features capture the perceived value and reputation of a car, which directly influence its price. Luxury brands like Mercedes-Benz may command higher prices compared to economy brands like Hyundai or Toyota.

4. **`type`**:
   - Different types of vehicles (e.g., SUVs, Sedans, Hatchbacks) cater to different market segments, influencing the price. SUVs often have higher prices compared to Sedans.

5. **`fuel` and `gearbox`**:
   - Fuel type (e.g., petrol vs. hybrid) and gearbox (e.g., manual vs. automatic) may influence the price based on buyer preferences and market demand.

6. **`color`**:
   - Although a less significant factor, popular colors (e.g., white, black) may slightly influence prices, especially for luxury vehicles.

7. **`condition`**:
   - The condition of the car (new vs. used) is a crucial factor, as new cars aren colors (e.g., white, black, grey, red) or 'others'         |

```python
from sklearn.model_selection import KFold
df_car_model = pd.read_csv('cleaned_car.csv')

df_car_model = df_car_model.drop(columns=['id', 'list_id', 'list_time'])

df_car_model = df_car_model[df_car_model['condition'] == 'used']


brand_distribution = df_car_model['brand'].value_counts(normalize=True)
cumulative_distribution = brand_distribution.cumsum()
selected_brands = cumulative_distribution[cumulative_distribution <= 0.9].index
df_filtered = df_car_model[df_car_model['brand'].isin(selected_brands)]

categorical_cols = ['brand', 'model', 'origin', 'type', 'gearbox', 'fuel', 'color', 'condition']
df_filtered = pd.get_dummies(df_filtered, columns=categorical_cols, drop_first=True)

X = df_filtered.drop(columns=['price'])
y = df_filtered['price']

kf = KFold(n_splits=5, shuffle=True, random_state=42)

```

```python
results = {
    "Model": [],
    "Version": [],
    "MSE": [],
    "R²": []
}

# Add entries for each model and its variants
def add_result(model_name, version, mse, r2):
    results["Model"].append(model_name)
    results["Version"].append(version)
    results["MSE"].append(mse)
    results["R²"].append(r2)
```

```python
def compute_f1_scores(clf, X, y, k=5):
    """Compute F1 scores using k-fold cross-validation."""
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    f1_macro_scores = cross_val_score(clf, X, y, cv=kf, scoring='f1_macro')
    return np.mean(f1_macro_scores)

def compute_auc_roc_scores(clf, X, y, k=5):
    """Compute AUC-ROC scores using k-fold cross-validation."""
    kf = KFold(n_splits=k, shuffle=True, random_state=42)
    y_probas = cross_val_predict(clf, X, y, cv=kf, method='predict_proba')
    auc_roc_score = roc_auc_score(y, y_probas[:, 1])  # Assuming positive class is at index 1
    return auc_roc_score

def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Greens', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.title('Confusion Matrix')
    plt.show()

def display_metrics(y_true, y_pred, y_proba):
    accuracy = accuracy_score(y_true, y_pred)
    auc_roc = roc_auc_score(y_true, y_proba)
    f1_macro = f1_score(y_true, y_pred, average='macro')
    precision = precision_score(y_true, y_pred, average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    print("Validation Accuracy: {:.3f}".format(accuracy))
    print("Validation AUC-ROC: {:.3f}".format(auc_roc))
    print("Validation F1-Score (Macro): {:.3f}".format(f1_macro))
    print("Validation Precision (Macro): {:.3f}".format(precision))
    print("Validation Recall (Macro): {:.3f}".format(recall))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import cross_val_score, KFold

# K-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mse_rf = []
r2_rf = []
fold = 1

for train_index, test_index in kf.split(X):
    print(f"\nFold {fold}")
    
    # Split the data into train and test for this fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Scale the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Train the Random Forest Regressor
    rf_model = RandomForestRegressor(
        max_depth=8, 
        n_estimators=500, 
        random_state=42
    )
    rf_model.fit(X_train_scaled, y_train)
    
    # Make predictions
    y_pred = rf_model.predict(X_test_scaled)
    
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"  Fold {fold} MSE: {mse:.4f}")
    print(f"  Fold {fold} R²: {r2:.4f}")
    
    mse_rf.append(mse)
    r2_rf.append(r2)
    fold += 1

# Average Performance Across All Folds
print("\nOverall Performance Across Folds:")
print(f"  Average MSE: {np.mean(mse_rf):.4f}")
print(f"  Average R²: {np.mean(r2_rf):.4f}")


# Baseline Random Forest
add_result("Random Forest", "Baseline", np.mean(mse_rf), np.mean(r2_rf))
```

## Finetuned Random Forest

```python

# Define the parameter grid for fine-tuning
param_grid = {
    'n_estimators': [100, 300, 500],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2', None]
}

# Define K-Fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Initialize the Random Forest Regressor
rf_model = RandomForestRegressor(random_state=42)

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, 
                           scoring='r2', cv=kf, verbose=2, n_jobs=-1)

# Fit the model with GridSearch
grid_search.fit(X, y)

# Get the best model and hyperparameters
best_rf_model = grid_search.best_estimator_
best_params = grid_search.best_params_
print("\nBest Hyperparameters:")
print(best_params)

# Evaluate the Best Model with K-Fold
mse_tuned_rf = []
r2_tuned_rf = []
fold = 1

for train_index, test_index in kf.split(X):
    print(f"\nFold {fold}")
    
    # Split the data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Train the fine-tuned model
    best_rf_model.fit(X_train, y_train)
    
    # Make predictions
    y_pred = best_rf_model.predict(X_test)
    
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"  Fold {fold} MSE: {mse:.4f}")
    print(f"  Fold {fold} R²: {r2:.4f}")
    
    mse_tuned_rf.append(mse)
    r2_tuned_rf.append(r2)
    fold += 1

# Overall Performance
print("\nOverall Performance Across Folds:")
print(f"  Average MSE: {np.mean(mse_tuned_rf):.4f}")
print(f"  Average R²: {np.mean(r2_tuned_rf):.4f}")

# Fine-Tuned Random Forest
add_result("Random Forest", "Fine-Tuned", np.mean(mse_tuned_rf), np.mean(r2_tuned_rf))
```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
import numpy as np

# Initialize variables to store results
kf = KFold(n_splits=5, shuffle=True, random_state=42)
baseline_mse_scores = []
baseline_r2_scores = []
fold = 1

# Perform K-Fold Cross-Validation for Baseline
for train_index, test_index in kf.split(X):
    print(f"\nFold {fold} - Baseline Gradient Boosting")
    
    # Split the data into train and test for this fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Scale the features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Train the Gradient Boosting Regressor (Baseline)
    baseline_gb = GradientBoostingRegressor(random_state=42)
    baseline_gb.fit(X_train_scaled, y_train)
    
    # Make predictions
    y_pred = baseline_gb.predict(X_test_scaled)
    
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"  Fold {fold} MSE: {mse:.4f}")
    print(f"  Fold {fold} R²: {r2:.4f}")
    
    baseline_mse_scores.append(mse)
    baseline_r2_scores.append(r2)
    fold += 1

# Average Performance Across All Folds for Baseline
avg_baseline_mse = np.mean(baseline_mse_scores)
avg_baseline_r2 = np.mean(baseline_r2_scores)

print("\nOverall Baseline Performance Across Folds:")
print(f"  Average MSE: {avg_baseline_mse:.4f}")
print(f"  Average R²: {avg_baseline_r2:.4f}")

add_result("Gradient Boosting", "Baseline", avg_baseline_mse, avg_baseline_r2)
```

## Fine-tuned Gradient Boosting

## Custom Neural Network (Tensorflow)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import KFold
import numpy as np

# K-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
mse_scores = []
r2_scores = []

# Initialize scaler for target variable
scaler_y = StandardScaler()

fold = 1
for train_index, test_index in kf.split(X):
    print(f"\nFold {fold}")
    
    # Split and scale data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1, 1))
    y_test_scaled = scaler_y.transform(y_test.values.reshape(-1, 1))
    
    # Define the model
    nn_model = Sequential([
        Input(shape=(X_train_scaled.shape[1],)),
        Dense(128, activation='relu'),
        BatchNormalization(),
        Dropout(0.2),
        Dense(64, activation='relu'),
        Dense(1)  # Output layer
    ])
    nn_model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')

    # Train the model
    nn_model.fit(X_train_scaled, y_train_scaled, epochs=100, batch_size=32, verbose=1)

    # Make predictions and inverse scale them
    y_pred_scaled = nn_model.predict(X_test_scaled).flatten()
    y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))

    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"  Fold {fold} MSE: {mse:.4f}")
    print(f"  Fold {fold} R²: {r2:.4f}")

    mse_scores.append(mse)
    r2_scores.append(r2)
    fold += 1

# Final results
print("\nFinal Neural Network Results:")
print(f"  Average MSE: {np.mean(mse_scores):.4f}")
print(f"  Average R²: {np.mean(r2_scores):.4f}")
```

```python
add_result("Neural Network", "Baseine", np.mean(mse_scores), np.mean(r2_scores))
```

## Finetuned Custom Neural Network

```python
# Initialize variables for Fine-Tuned Model
fine_tuned_mse_scores = []
fine_tuned_r2_scores = []
fold = 1

# Perform K-Fold Cross-Validation for Fine-Tuned Model
for train_index, test_index in kf.split(X):
    print(f"\nFold {fold} - Fine-Tuned Neural Network")
    
    # Split the data into train and test for this fold
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Scale the features
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)
    
    # Define the Fine-Tuned Model
    fine_tuned_nn = Sequential([
        Input(shape=(X_train_scaled.shape[1],)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(32, activation='relu'),
        Dense(1)  # Output layer
    ])
    fine_tuned_nn.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    
    # Train the model
    fine_tuned_nn.fit(X_train_scaled, y_train, epochs=100, batch_size=16, verbose=0)
    
    # Make predictions
    y_pred = fine_tuned_nn.predict(X_test_scaled).flatten()
    
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    print(f"  Fold {fold} MSE: {mse:.4f}")
    print(f"  Fold {fold} R²: {r2:.4f}")
    
    fine_tuned_mse_scores.append(mse)
    fine_tuned_r2_scores.append(r2)
    fold += 1

# Average Performance Across All Folds for Fine-Tuned Model
avg_fine_tuned_mse = np.mean(fine_tuned_mse_scores)
avg_fine_tuned_r2 = np.mean(fine_tuned_r2_scores)

print("\nOverall Fine-Tuned Performance Across Folds:")
print(f"  Average MSE: {avg_fine_tuned_mse:.4f}")
print(f"  Average R²: {avg_fine_tuned_r2:.4f}")

add_result("Neural Network", "Fine-tuned", avg_fine_tuned_mse, avg_fine_tuned_r2)
```
